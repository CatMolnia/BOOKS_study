"""---1.3.2. Операции над объектами DataFrame---"""

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv', index_col='Title') # импорт данных
```

```python
movies.iloc[499] # извлекаем строку по индексу (числовой позиции)
```

```python
movies.loc['Forrest Gump'] # извлекаем строку по индексу (по метке индекса)
```

```python
movies.sort_values(by='Year', ascending=False).head() # сортируем датафрэйм по одному признаку
```

```python
movies.sort_values(by=['Studio', 'Year']).head() # сортируем датафрэйм по нескольким признакам
```

```python
movies.sort_index().head() # сортируем индексы
```

"""---1.3.3. Подсчет значений в Series---"""

```python
movies['Studio'].value_counts().head(10) # считаем количество значений в указанном столбце
```

"""---1.3.4. Фильтрация столбца по одному или нескольким критериям---"""

```python
movies[movies["Studio"] == "Universal"] # фильтруем указанный столбец по условию
```

или

```python
released_by_universal = (movies["Studio"] == "Universal")
movies[released_by_universal].head()
```

```python
released_by_universal = movies["Studio"] == "Universal"
released_in_2015 = movies["Year"] == 2015
movies[released_by_universal & released_in_2015] # фильтрация по нескольким признакам (и то и другое (&))
```

```python
released_by_universal = movies["Studio"] == "Universal"
released_in_2015 = movies["Year"] == 2015
movies[released_by_universal | released_in_2015] # фильтрация по нескольким признакам (или то или другое (|))
```

```python
before_1975 = movies['Year'] < 1975
movies[before_1975]
```

```python
mid_80 = movies['Year'].between(1983, 1986) # фильтрация в интервале
movies[mid_80]
```

```python
has_dark_in_title = movies.index.str.lower().str.contains("dark")
movies[has_dark_in_title]
```
